```python
# Data Handling
import pandas as pd
import numpy as np

# Encoding
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

# Feature Scaling
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

# Train Test Split
from sklearn.model_selection import train_test_split

# Column Transformer
from sklearn.compose import ColumnTransformer

# Pipeline
from sklearn.pipeline import Pipeline
```

```python
data = pd.read_excel("../dataset/cleaned_output.xlsx")
```

```python
le = LabelEncoder()

data["Country_Label"] = le.fit_transform(data["Country"])

print(data[["Country", "Country_Label"]].head())
```

```python
country_encoded = pd.get_dummies(
    data["Country"],
    prefix="Country"
)

data = pd.concat(
    [data, country_encoded],
    axis=1
)

data.head()
```

```python
data["Revenue"] = data["Quantity"] * data["UnitPrice"]
```

```python
features_to_scale = [
    "Quantity",
    "UnitPrice",
    "Revenue",
    "Country_Label"
]
```

```python
scaler = StandardScaler()

data[features_to_scale] = scaler.fit_transform(
    data[features_to_scale]
)

data.head()
```

```python
scaler = MinMaxScaler()

data[features_to_scale] = scaler.fit_transform(
    data[features_to_scale]
)
```

```python
X = data[features_to_scale]

print(X.head())
```

```python
X_train, X_test = train_test_split(
    X,
    test_size=0.20,
    random_state=42
)

print("Training Shape :", X_train.shape)
print("Testing Shape  :", X_test.shape)
```

```python
X_train.to_csv("../dataset/X_train.csv", index=False)
X_test.to_csv("../dataset/X_test.csv", index=False)
```
