```python
import pandas as pd
import numpy as np

# Clustering Models
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN

from sklearn.preprocessing import StandardScaler
# Evaluation
from sklearn.metrics import silhouette_score

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

import os
import joblib
```

```python
X_train = pd.read_csv("../dataset/X_train.csv")
X_test = pd.read_csv("../dataset/X_test.csv")
```

```python
X_train_small = X_train.sample(n=1000, random_state=42)
X_test_small = X_test.sample(n=1000, random_state=42)
```

```python
X_train_small.to_csv("../dataset/X_train_1k.csv", index=False)
X_test_small.to_csv("../dataset/X_test_1k.csv", index=False)
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train_small)
```

```python
# ==========================
# K-Means Clustering
# ==========================

kmeans = KMeans(
    n_clusters=4,
    random_state=42,
    n_init=10
)

kmeans.fit(X_train_scaled)

kmeans_labels = kmeans.predict(X_train_scaled)

kmeans_score = silhouette_score(
    X_train_scaled,
    kmeans_labels
)

print("KMeans Silhouette Score:", kmeans_score)
```

```python
# ==========================
# Hierarchical Clustering
# ==========================

hierarchical = AgglomerativeClustering(
    n_clusters=4,
    linkage="ward"
)

hierarchical_labels = hierarchical.fit_predict(
    X_train_scaled
)

hierarchical_score = silhouette_score(
    X_train_scaled,
    hierarchical_labels
)

print(
    "Hierarchical Silhouette Score:",
    hierarchical_score
)
```

```python
# ==========================
# DBSCAN
# ==========================

dbscan = DBSCAN(
    eps=0.5,
    min_samples=10
)

dbscan_labels = dbscan.fit_predict(
    X_train_scaled
)

# Remove noise points (-1)
mask = dbscan_labels != -1

if len(set(dbscan_labels[mask])) > 1:

    dbscan_score = silhouette_score(
        X_train_scaled[mask],
        dbscan_labels[mask]
    )

    print(
        "DBSCAN Silhouette Score:",
        dbscan_score
    )

else:
    print(
        "DBSCAN produced insufficient clusters."
    )
```

```python
# ==========================
# Save Models
# ==========================

os.makedirs("models", exist_ok=True)

joblib.dump(
    kmeans,
    "../models/kmeans_model.pkl"
)

joblib.dump(
    hierarchical,
    "../models/hierarchical_model.pkl"
)

joblib.dump(
    dbscan,
    "../models/dbscan_model.pkl"
)

joblib.dump(
    scaler,
    "../models/scaler.pkl"
)

joblib.dump(
    X_train.columns,
    "../models/columns.pkl"
)

print("All models saved successfully.")
```
